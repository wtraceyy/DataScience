```python

def get_grade(score):
    if score >= 80:
        return 'A'

    elif score >= 70:
        return 'B'

    elif score >= 60:
        return 'C'

    elif score >= 50:
        return 'D'

    elif score >= 40:
        return 'E'

    else:
        return 'F'

scores = [78, 49, 67, 89, 43, 50, 86, 34]

for score in scores:
    grade = get_grade(score)

    if score >= 50:
        status = 'Pass'
    else:
        status = 'Fail'

    print(score, grade, status)
```
