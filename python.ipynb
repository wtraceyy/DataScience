```python
#EXERCISE 3

students = [
    {
        "name": "Alice",
        "scores": [80, 75, 90],
        "city": "Nairobi"
    },

    {
        "name": "Brian",
        "scores": [60, 55, 70],
        "city": "Mombasa"
    },

    {
        "name": "Cynthia",
        "scores": [95, 88, 92],
        "city": "Kisumu"
    },

    {
        "name": "David",
        "scores": [40, 50, 45],
        "city": "Nairobi"
    },

    {
        "name": "Esther",
        "scores": [78, 85, 80],
        "city": "Eldoret"
    }
]

# Function to calculate average
def get_average(scores):
    return sum(scores) / len(scores)


# Function to calculate grade
def get_grade(average):

    if average >= 90:
        return "A"

    elif average >= 80:
        return "B"

    elif average >= 70:
        return "C"

    elif average >= 60:
        return "D"

    else:
        return "F"


# Print student report
print("STUDENT REPORT")
print("-" * 30)

for student in students:

    try:
        average = get_average(student["scores"])
        grade = get_grade(average)

        print(f"Name: {student['name']}")
        print(f"Scores: {student['scores']}")
        print(f"Average: {average:.2f}")
        print(f"Grade: {grade}")
        print(f"City: {student['city']}")
        print("-" * 30)

    except ValueError:
        print("Invalid score detected")


# Sort students by average score (highest first)
sorted_students = sorted(
    students,
    key=lambda student: get_average(student["scores"]),
    reverse=True
)

print("\nSTUDENT RANKING")
print("-" * 30)

for student in sorted_students:
    avg = get_average(student["scores"])
    print(f"{student['name']} - {avg:.2f}")


# Count unique cities using a set
cities = set()

for student in students:
    cities.add(student["city"])

print("\nUnique Cities:", cities)
print("Number of unique cities:", len(cities))
```

```python
#DECORATORS AND PROPERTIES

#Decorators
#it is a function upgrader
#a function that takes a function in abd returns an imporved versiom
#denoted using "@"
#A decorator modifies or extends a function without changing its code.

def shout(func):                        #takes function in
    def wrapper():
        return func().upper() +"!!"
    return wrapper                      #returns an upgraded function

@shout                    #Short hand for : greet = shout
def greet():
    return "HELLO!!"   #HELLO!! :greet was replaced by the upgraded version



#How it works
#say_hello() normally just prints "Hello!".
#The decorator adds extra code before and after it runs.
#So when you call say_hello(), you get:

def my_decorator(func):
    def wrapper():
        print("Before function")
        func()
        print("After function")
    return wrapper


@my_decorator
def say_hello():
    print("Hello!")

say_hello()


#Properties
#is a fake attribute
#it looks like a variable from the outside but is sceretly a function running on the inside
# a normal attribute is like a box : reading it fetches the stored value
# a property has no box ; it runs a geeter and hands you wtv it returns

class Person:
    def __init__(self, age):
        self._age = age

    @property
    def age(self):
        return self._age

p = Person(20)
print(p.age)

#Properties with validation
class Person:
    def __init__(self, age):
        self.age = age

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value
        

#Decorators and properties combined
class Student:
    def __init__(self, score):
        self._score = score

    @property
    def score(self):
        return self._score

    @score.setter
    def score(self, value):
        if value < 0 or value > 100:
            raise ValueError("Score must be between 0 and 100")
        self._score = value

    s = Student(80)
    print(s.score)
    s.score = 95
```
