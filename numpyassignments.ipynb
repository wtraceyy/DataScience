### HANDLING NaN

```python
sales = daily_sales.astype(float)

sales[2,1] = np.nan
sales[5,3] = np.nan
sales[8,0] = np.nan

print(sales)
```

```python
missing = np.isnan(sales)
print(missing)

print("Total missing values:")
print(missing.sum())

print("Missing values in each column:")
print(missing.sum(axis=0))
```

```python
print("Mean ignoring NaNs:")
print(np.nanmean(sales,axis=0))

print("Normal mean:")
print(sales.mean(axis=0))
```

```python
medians = np.nanmedian(sales,axis=0)

for column in range(sales.shape[1]):
    missing = np.isnan(sales[:,column])
    sales[missing,column] = medians[column]

print(sales)
```

```python
def clean_array(arr):
    arr = arr.astype(float)
    arr[arr==-1] = np.nan

    means = np.nanmean(arr,axis=0)

    for column in range(arr.shape[1]):
        mask = np.isnan(arr[:,column])
        arr[mask,column] = means[column]

    return arr
```
