### Measures of Dispersion(Spread)
These show how spread out the data is.

**Range**
Difference between the highest and lowest values.
Range=Maximum−Minimum

**Variance**
Measures how far values are from the mean on average.
Variance = Average squared distance from the mean
POPULATION VARIANCE
variance = sum((x - mean)^2) / N

SAMPLE VARIANCE
variance = sum((x - mean)^2) / (n - 1)

**Standard Deviation**
The square root of variance.

POPULATION STANDARD DEVIATION
std_dev = sqrt(sum((x - mean)^2) / N)

SAMPLE STANDARD DEVIATION
std_dev = sqrt(sum((x - mean)^2) / (n - 1))

A small standard deviation means data points are close to the mean.
A large standard deviation means data points are spread out.

x      = each data value
mean   = average of the data
N      = population size
n      = sample size
sqrt() = square root

### Types of events
**Mutually exclusive** → cannot happen together
Two events cannot happen at the same time.
Getting Head or Tail in one coin toss

**Independent** → no influence
One event does not affect the other.
Tossing a coin and rolling a die

**Dependent** → influence each other
One event affects the probability of another.
Picking cards without replacement

**Complementary** → A and NOT A
Two events where one must happen if the other does not
Getting a Head vs NOT getting a Head

**Collectively exhaustive** → cover all outcomes

### Probability rules
(a) Complement Rule
Probability that event does NOT happen.
P(A') = 1 - P(A)

(b) Addition / OR Rule
Used when we want A OR B.
(non mutually exclusive) P(A ∪ B) = P(A) + P(B) - P(A ∩ B)
Special case (mutually exclusive): P(A ∪ B) = P(A) + P(B)

(c) Multiplication Rule
Used for AND events.
Independent events: P(A ∩ B) = P(A) × P(B)
Dependent events: P(A ∩ B) = P(A) × P(B|A)

(d) Conditional Probability
Probability of A happening GIVEN B has happened.
P(A|B) = P(A ∩ B) / P(B)

### 🧠Bayesian Thinking 
Framework of reasoning against uncertainity

### Bayes’Theorem (Bayes’ Rule)
used to claculate the conditional probability of an event based on the occurrence of another event
a mathematical rule for updating a belief when new evidence arrives
P(A|B) → probability of A given B has happened (posterior probability)
P(B|A) → probability of B given A is true (likelihood)
P(A) → prior probability (before evidence)
P(B) → total probability of B (normalizing factor/evidence)

P(A|B) = P(B|A) * P(A)
         -------------
              P(B)





```python
# A disease affects 1 in 1000 people .A test is 99% accurate .You test positive 
#What's the chance you're actually sick .
# Intuition screams 99% . Bayes says = 9% why? because the disease is so rare thar rhe few false positives from the huge healthy population vastly outnumber the true positives from the tiny sick population
#Your prior(rarity) dominates. 
#This is the most important counter- intuitive result in applied statistics and it's why doctors retest before diagnosing

# MEDICAL TEST
p_disease = 0.001 #prior probability
p_pos_given_disease = 0.99 #likelihood  
p_pos_given_healthy = 0.01 #false positive rate

#Total probability of testing positive (the evidence,p(b))
p_pos = (p_pos_given_disease * p_disease + p_pos_given_healthy * (1-p_disease))

#Posterior : p(disease|positive)
posterior = (p_pos_given_disease * p_disease) / p_pos

print(round(posterior , 4))

tp = p_pos_given_disease * p_disease
fp = p_pos_given_disease *(1- 0.99)

probability = tp / (tp + fp)
print(f"Chance you're actually sick: {probability:.2%}")

```

```python
import numpy as np
from scipy.stats import binom

n = 25
p = 0.7

# 1. Expected value (Built-in method instead of n * p)
expected = binom.mean(n, p)
print(f"Expected full trips: {expected}")

# 2. Probability of at least 20 full trips
# "At least 20" means we want to subtract "19 or fewer" from 100% (1)
prob_at_least_20 = 1 - binom.cdf(19, n, p)
print(f"P(X >= 20): {prob_at_least_20:.4f}")


from scipy.stats import poisson

# Part A: Exactly 6 calls in 1 hour (Average rate = 4)
prob_6 = poisson.pmf(6, mu=4)
print(f"P(exactly 6 calls): {prob_6:.4f}")

# Part B: None in 15 minutes (Average rate scales down to 1)
prob_none = poisson.pmf(0, mu=1)
print(f"P(0 calls in 15 mins): {prob_none:.4f}")

```

### CDF

### HYPOTHESIS TESTING
Hypothesis testing is a statistical method used to decide whether there’s enough evidence in data to support a claim about a population.

It is a formal procedure for deciding whether a pattern in our sample is real enough to believe about the population or whether it could easily be an error of random sampling

1. Null Hypothesis (H0)
This is the “default” assumption.

Example:
“This new medicine has no effect.”
“The average height is 170 cm.”

2. Alternative Hypothesis (Ha or H1)
This is what you want to test for.

Example:
“The medicine does work.”
“The average height is not 170 cm.”

### Steps in Hypothesis Testing
**Step 1: State hypotheses**
Example:

H0:μ=170
Ha:μ=170

**Step 2: Collect data**
Take a sample and calculate statistics like the sample mean.

**Step 3: Choose significance level (α)**
Usually: α=0.05
This means you allow a 5% chance of wrongly rejecting the null hypothesis.

**Step 4: Compute a test statistic**
Examples:
        z-score
        t-score
        chi-square

This measures how far your sample result is from what the null hypothesis predicts.

**Step 5: Find the p-value**
The p-value tells you:
“If the null hypothesis were true, how likely is it to get results this extreme?

**Step 6: Make a decision**
If p-value ≤ α → Reject H0
If p-value > α → Fail to reject H0
	​
### Types of errors
**Type I error**
Also called false positive
Occurs when you rejext the null hypothesis when it is actually true

**Type II error**
Also called false negative

### The Central Limit Theorem (CLT)
It says if you take any sample and compute each one's mean , the distribution of those sample means is approximately normal regardless of the shape of the original data as long as the sample size is reasonably large (n >= 30)

It explains why the normal distribution (bell curve) appears everywhere.

Individual matatu fares might be wildly skewed (that 250 outlier)but the average fare of 40 random trips is a much steadier thing extreme highs and lows tend to cancel within each sample . do that averaging many times and the averages pile up into a neat bell curve centered on the true population mean


### Standard Error vs Standard Deviation
Standard Error (SE)
Measures how spread of sample mean is across repeated samples
SE = (​σ/sqrt​ n)

Standard Deviation (σ)
Measures spread of individual observations.

Student scores:
40,60,80,100
The SD describes how spread out the students are.

```python
import numpy as np, matplotlib.pyplot as plt
population = np.random.exponential(scale = 70 , size = 100000)

means = [np.random.choice(population , 40).mean() for _ in range(1000)]
plt.hist(means, bins = 30)
plt.title("CLT : distribution of sample mean is normal")
plt.show

fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].hist(population, bins=50)
ax[0].set_title("Population (Exponential)")

ax[1].hist(means, bins=30)
ax[1].set_title("Sample Means (n=40)")

plt.show()
```
