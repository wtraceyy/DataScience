```python
#EXERCISE ON POLYMORPHISM

class Chart:
    def __init__(self, title , data):

        if not isinstance(data,list) or len(data) == 0:
            raise ValueError('Data must be a non-empty list')
        
        self.title = title
        self.data = data
        
    def render(self):
        return f"{self.title} Rendering a generic chart {self.data}"
    
    def __str__(self):
        return f"Generic Chart:{self.title}"

class BarChart(Chart):
    def render(self):                               
        return f"{self.title} Formatted bar chart description {self.data}"
    
    def __str__(self):
        return f"BarChart Object:{self.title}"
    
class LineChart(Chart):
    def render(self):                               
        return f"{self.title} Linechart description {self.data}"
    
    def __str__(self):
        return f"LineChart Object:{self.title}"
     
class PieChart(Chart):
    def render(self):                               
        return f"{self.title} Piechart description {self.data}"
    
    def __str__(self):
        return f"PieChart Object:{self.title}"
    
class ScatterPlot:
    def __init__(self, title , data):
        
        if not isinstance(data,list) or len(data) == 0:
            raise ValueError('Data must be a non-empty list')
        
        self.title = title 
        self.data = data

    def render(self):
        return f"Scatterplot {self .title} Rendering a scatter plot method using duck typing with data{self.data}"

    def __str__(self):
        return f"ScatterPlot Object:{self.title}"
    
dashboard = [Chart('Maths',[60]),
             BarChart('Eng',[77]),
             PieChart('Sci',[48]),
             LineChart('Geo',[29]),
             ScatterPlot('Phy',[64])
             ]

def render_all(Charts):
    for i in Charts:
     print(i.render())


render_all(dashboard)
     
```

```python
#TASKS

class Student:
    def __init__(self, name, age, scores):
        self.name = name
        self.age = age
        self.scores = scores

    def get_average(self):
        average = sum(self.scores) / len(self.scores)
        return average
        
    def get_grade(self):
        average = self.get_average()

        if average >= 80:
           return "A"
        elif average >= 70:
            return "B"
        elif average >= 60:
           return "C"
        elif average >= 50:
            return "D"
        else:
            return "F"
        
    
    def full_report(self):
        return f"Name {self.name}, Age{self.age}, Number of exams{len(self.scores)} ,Average{self.get_average} ,Grade{self.get_grade}" 
    
school_name = "Data Science Academy"      
    
def __str__(self):
   return f"Student: {self.name} ,Age: {self.age}, Avg: {self.get_average():.1f})"

def add_score(self, score):
        if not (0 <= score <= 100):
            raise ValueError("Score must be between 0 and 100.")
        self.scores.append(score)

def is_improving(self):
        if len(self.scores) < 2:
            return False
        return self.scores[-1] > self.scores[0]

        
classroom = [Student('Tracey',18 ,[85, 92, 88]),
             Student('Jeremy',21 ,[70, 65, 74]),
             Student('Rue',19 ,[95, 98, 100]),
             Student('Timothy',18 ,[60, 75, 82]),
             Student('Jane',20,[74, 49, 43])

]

print(f"Age of student: {classroom[0].age}")

def full_report(classroom):
    for student in classroom:
        student.full_report()

highest_student = classroom[0]

for student in classroom:
    current_avg = student.get_average() 
    highest_avg = highest_student.get_average()
    
    if current_avg > highest_avg:
        highest_student = student

print("Top Performing Student:")
print(highest_student.name)

    
```
