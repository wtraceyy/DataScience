### CALCULUS
Calculus is a branch of mathematics that studies change and motion. It helps us understand how things grow, shrink, move and accumulate over time.

**Why is calculus important?**
Calculus is used in many fields, including:
Physics (motion, forces, electricity)
Engineering (designing bridges, aircraft, and electronics)
Economics (optimizing profits and costs)
Biology and medicine (modeling population growth and disease spread)
Computer science (machine learning, graphics, and optimization)

**Why is it the core of machine learning**
Calculus is considered the core of machine learning because machine learning is essentially about learning from mistakes. Calculus provides the mathematical tools that tell a model how to improve

**Differential calculus**
Studies rates of change/slope of function at a given point.
Tells you how fast something is changing at any moment.

**Integral calculus**
Studies accumulation.
How much has piled over time

### SUPPORT VECTOR MACHINES
Combines ideas from linear algebra ,calculus ,geometry and optimization

**Problem it tries to solve**
Find a line or plane that separates two classes
Many lines can separate them but which one should we choose

**The best separating class**
Separate them as far away from both as possible
leads to the maximum margin principle

**The Margin**
This is the distance between the decision boundary and the closest point
The decision boundary is where predictions change
This is the perpendicular distance from the separating hyperplane to the nearest training examples on either side.

*Why maximize the margin?*
A larger margin means the classifier is less sensitive to small changes in the data.
This improves generalization.

**Hyperplane**
A line separates points in 2D.
