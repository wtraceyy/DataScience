### What is linear algebra
Linear algebra is the branch of mathematics that studies:
Vectors                      Linear equations
Transformations of data      Matrices

### Vectors
An ordered list of numbers that represents a point in space
The order matters i.e each number in the list is a competent
The dimension of a vector is simply the count of its component

# Types of vectors
Row vectors ;written horizontally
A = [2,3,4]
Column vectors ;written vertically

```python
import numpy as np
# A 1D numpy array is our basic vector
matatu = np.array([8, 12, 3.5, 120, 0])
print('Shape:',matatu.shape)

#column vector(2D array with one column )
col = matatu.reshape(-1 , 1)
print("Column shape:",col.shape)
```
