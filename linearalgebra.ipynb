### INVERSE OF A MATRIX
if determinant is 1 it is invertible
if determinant is 0 it is not invertible

The inverse of a matrix is similar to the reciprocal of a number.
A × A−1 = I
where I is the identity matrix.

**When does an inverse NOT exist?**
When the determinant is 0  Why?
The rows are linearly dependent.
When the matrix is not a square matrix
One equation is just a copy of the other, so the matrix doesn't contain enough independent information.
Matrices with determinant zero are called singular matrices.

**Effects of a non-invertible matrix(singular matrix)**
A non-invertible matrix (also called a singular matrix) is a matrix that has no inverse. This happens when its determinant is zero: det(A)=0
You cannot reverse the transformation
Information is lost
Linear systems may have no unique solution
Columns become linearly dependent
Rank becomes smaller than the matrix size

### PRINCIPLE COMPONENT ANALYSIS
PCA (Principal Component Analysis) is a technique that simplifies data by reducing the number of variables (features) while preserving as much important information as possible.

**Why is it called "Principal Component Analysis"?**
Principal = most important.
Component = a new variable created from the original variables.
Analysis = the process of finding these components.

**Why do we use PCA? / Importance**
Reduce the number of features (dimensionality reduction).
Remove redundant information.
Make machine learning models faster and sometimes more accurate.
Visualize high-dimensional data (e.g., reducing 100 features to 2 so you can plot it).

**Core Idea of PCA**
PCA finds:
First Principal Component
The direction with the largest variance.
Meaning:
the direction where data spreads out the most.

Second Principal Component
The next most important direction.

Requirements:
captures maximum remaining variance
perpendicular to the first component

Third Component
Perpendicular to the first two.

**Variance is the Key**
PCA assumes:
High variance = important information
and
Low variance = mostly noise

**Step-by-Step PCA**
Step 1: Standardize Data
Variables may have different scales.
z = x−μ
     σ
	
Step 2: Create Covariance Matrix​
Covariance tells us whether variables move together.

Step 3: Compute Eigenvalues and Eigenvectors
Eigenvalue = variance explained.
Find eigenvalues
Find eigenvectors
Sort eigenvalues largest → smallest

Step 4: Choose Principal Components
If first principle component is enough , discard the second one

Step 5: Transform Data
Project data onto selected eigenvectors.
Formula: Z = XW

where
X = original data
W = chosen eigenvectors

The result is the PCA-transformed dataset.

**Relationship Between Eigenvalues and PCA**
Think of PCA as asking:
"Which directions matter most?"
Eigenvectors answer:
"These are the directions."
Eigenvalues answer:
"This is how important each direction is."
PCA = finds those directions, ranks them by importance, and keeps only the most important ones.
