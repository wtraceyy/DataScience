## HANDLING MISSING VALUES
### **Flagging Missing Values**
It creates an indicator showing which values were originally missing.
Instead of only filling the missing values, you create another column showing where data was originally missing.
df["Age_missing"] = df["Age"].isnull()
df["Age_missing"] = df["Age"].isna()

*Why flag missing values?*
Sometimes missing values are meaningful.

### **Interpolation**
Interpolation is the process of estimating missing values based on the values around them.
Interpolation assumes that the missing value follows the trend of the surrounding data.

#### Types of interpolation
1. Linear interpolation
Assumes data changes in a straight line.

2. Polynomial interpolation
Uses a curved line rather than a straight line.
Useful when data naturally curves.

3. Time interpolation
Useful for time series.

### **Replacing Missing Values/Imputation**
Means filling NaN values with something else.

You can replace with:
                     Mean
                     Median
                     Mode
                     Constant value
                     Previous value
                     Next value
                     Predicted value from a machine learning model

###  **Simple Imputer**
SimpleImputer is a class from scikit-learn that automatically replaces missing values using a chosen strategy.
Automates imputation using strategies like mean, median, mode, or a constant

*Why use SimpleImputer?*
It integrates well into machine learning workflows, especially when building preprocessing pipelines.

### **Forward Fill (ffill)**
Forward fill copies the previous valid value into the missing value.
df.ffill()
df.fillna(method="ffill")

*Advantages*
Very fast
Good for time series data
Preserves previous observations

*Disadvantages*
Can repeat outdated values if there are many consecutive missing entries

### **Backward Fill (bfill)**
Backward fill copies the next valid value backward into missing entries.
Use the next available value.
df.bfill()
df.fillna(method="bfill")

*Advantages*
Useful when later observations are more appropriate than earlier ones
Common in some time series or ordered datasets

*Disadvantages*
May use future information, which is not always appropriate (for example, in forecasting tasks)

```python
# Detecting the missing value
# Missing values in pandas show up as NaN
# We find them first then decide what to do

df.isna()
df.isna().sum() #Count of missing per column
df['salary_history_2022'].isna()

# filter rows where a value is missing
df[df['salary_history_2022'].isna()]

# filter rows where a value is not missing
df[df['salary_history_2022'].notna()]


```

```python
# Dropping missing values
# Use this when the rows or columns with missing data are not useful
  
df.dropna()           # drops any row with at least one missing value
df.dropna(how='all')  # drops only if the entire row is NaN
df.dropna(subset=['salary_history_2022'])  #drops rows where the columns is NaN
df.dropna(axis = 1)      #drops columns that have any NaN values
df.dropna(thresh = 8)    # keeps rows with at least 8 NaN values
```

```python

# Fill all missing values in the DataFrame with a placeholder
df = df.fillna('Unknown')

# Fill numeric column with the median
median_val = df['salary_history_2022'].median()
df['salary_history_2022'] = df['salary_history_2022'].fillna(median_val)

# Fill text column with the most common value (mode)
df['employment_type'] = df['employment_type'].fillna(df['employment_type'].mode()[0])

# Forward fill
df['salary_history_2022'] = df['salary_history_2022'].ffill()
# df['salary_history_2022'].fillna(method='ffill', inplace=True)

# Backward fill
df['salary_history_2022'] = df['salary_history_2022'].bfill()
# df['salary_history_2022'].fillna(method='bfill', inplace=True)
```

```python
#Interpolation
#For numeric columns in sequence , estimate the missing values
df['salary_history_2022'].interpolate(method='linear', inplace=True)

# Flagging that a value is missing
# One might want to keep the record that the value was originally missing

df['had_missing_salary']= df['salary_history_2022'].isna()
df['salary_history_2022'] = df['salary_history_2022'].fillna(df['salary_history_2022'].median())   
```

### Grouping and aggregation

#### Group by
It wrks in 3 steps 
split - divides the data into groups
apply - runs a function on each group
combine - brings the result together
'for each[category],calculate[something]'

```python
df.groupby('job_title')['salary_kes'].mean()  #average salary per role
df.groupby('city')['name'].count()            #headcount per city
df.groupby('employment_type')['salary_kes'].max()     
```

```python
# Common aggregation functions
df.groupby('job_title')['salary_kes'].mean()  #average
df.groupby('job_title')['salary_kes'].max     #highest value
df.groupby('job_title')['salary_kes'].min()   #lowest value
df.groupby('job_title')['salary_kes'].std()   #std deviation
df.groupby('job_title')['salary_kes'].count()  #non-null count
df.groupby('job_title')['salary_kes'].sum()  #average salary per role
df.groupby('job_title')['salary_kes'].median()  #middke
df.groupby('job_title')['salary_kes'].size()  #group size , all rows
```

#### Multiple Statistics at Once

```python
df.groupby('job_title')['salary_kes'].agg(['mean', 'median', 'min', 'max', 'count'])

# Different stats for different columns
df.groupby('job_title').agg({
 'salary_kes': ['mean', 'max'],
 'experience_yrs': ['mean', 'min'],
 'age': 'median'
})
```

#### Grouping by Multiple Columns

```python
df.groupby(['city', 'employment_type'])['salary_kes'].mean()
df.groupby(['job_title', 'city']).agg({
 'salary_kes': 'mean',
 'experience_yrs': 'mean'
})
```

### Merging and Combining Data Sets
Merging combines two DataFrames based on a shared column — exactly like SQL JOINs.

```python
departments = pd.DataFrame({
 'job_title': ['Data Analyst', 'Software Engineer', 'Data Scientist', 'ML Engineer', 
'DevOps Engineer'],
 'department': ['Analytics', 'Engineering', 'Analytics', 'AI/ML', 'Infrastructure'],
 'budget_kes': [5000000, 8000000, 6000000, 9000000, 7000000]
})
print(departments)

# Inner join — only rows where job_title exists in BOTH tables
merged = pd.merge(df, departments, on='job_title', how='inner')
print(merged)
```

```python
# The four types of joins
pd.merge(df, departments, on='job_title', how='inner') # only matching rows
pd.merge(df, departments, on='job_title', how='left') # all left + NaN for missing right
pd.merge(df, departments, on='job_title', how='right') # all right + NaN for missing left
pd.merge(df, departments, on='job_title', how='outer') # all rows from both
```
