### Indexing and Slicing
Indexing means grabbing a single element
Slicing means grabbing a subset


```python
# 1D indexing and slicing
# Here we give one index in the square brackets and it will return the entire row of that index
#Average sales per day
average_daily_sales = np.mean(mock_sales,axis = 0)
print(average_daily_sales)

#Access by position
print(average_daily_sales[0]) # first column
print(average_daily_sales[-1]) # last column
print(average_daily_sales[1:3]) # second and third column

#Slicing : syntax 
#Stop is exclusive - the element that index is not included in the slice
#Stop is optional - if not provided , it will default to one
print(average_daily_sales[:2]) #first two columns
print(average_daily_sales[::2]) # every other column
print(average_daily_sales[:-1]) #reverse the array
```

### 2D indexing and slicing
Here we need two indices: [row,column].The comma separates them
A colon by itself means "all of them"

```python
#single element : mock sales{row,column}
print(mock_sales)

print(mock_sales[0,0]) #first row, first column
print(mock_sales[0,1]) # first row , second column
print(mock_sales[-1,-1]) # last row , last column

#entire row: daily sales for the first day
print(mock_sales[0,:]) #first row , all columns

#entire column: daily sales for the first product
first_product_sales = mock_sales[:,0] 
print(first_product_sales)

#subgrid: first 3 days , first 2 products
print(mock_sales[:3,:2])

#last 3 days , last 2 products
print(mock_sales[-3:,-2:])

#entire other days , all categories
alternate_days_sales = mock_sales[-3:,-2:]
print(alternate_days_sales)
print(alternate_days_sales.shape)

#NB
#Slices are vies , not copies . If you modify a slice , use the copy() method. If you want an independent version ; subgrid 

#proof that slicing shares memory with the original array
view = mock_sales[:2, :]
view[0,0] = 9999
print(mock_sales[0,0])

#safer version using copy()
mock_sales[0,0] = 5000 #Resetting the value back to original for demonstration
safe_copy = mock_sales[:2,:].copy()
safe_copy[0,0] = 9999
print(mock_sales[0,0])
print(safe_copy[0,0])
print(mock_sales)
print(safe_copy)
```

### Aggregation and Axis Concepts
Aggregation collapses an array into one or more summary values
Total , average, minimums and so on
The axis parameter controls which direction you collapse

```python
# Basic Aggregation
# Sum of all values in the entire array

import numpy as np
#100 days of sales data across 4 categories
#rows = days , columns = categories

np.random.seed(42)
daily_sales = np.random.normal(loc=13000,scale=1500,size=(10,4)).round().astype(int)
print(daily_sales.shape)

#sum of all values in the entire array
print(daily_sales.sum())

#product of elements(useful for growth rates)
growth_rates = np.array([1.0, 1.02, 1.05, 0.98, 1.03])
compound = growth_rates.prod()
print(f"Compound effect: {compound:.4f}")

#cumulative sum-running total of food sales over 10 days
food_cumsum = daily_sales[:,0].cumsum()
print("Running food revenue:",food_cumsum)

#cumulative product
rates = np.array([1.0, 1.02, 1.05, 0.98, 1.03])
print("Compound after each period:",rates.cumprod(),round(4))
```

#### Statistical aggregation

```python
food_sales = daily_sales[:,0] #all 10 days of food

#Mean
print(f"Mean food sales :{food_sales.mean():.0f}")

#Median
print(f"Median food sales :{np.median(food_sales):.0f}")

#Standard deviation
print(f"Std dev:{food_sales.std():.0f}")

#Variance(std squared)
print(f"Variance:{food_sales.var():.0f}")

#Min and max
print(f"Lowest day:{food_sales.min()}")
print(f"Highest day:{food_sales.max()}")

print(f"Best day:")
```

#### Axis Concept
Axis controls which direction numpy collapses when it aggregates
axis = 0 collapses the row , one result per column , axis = 1 collapses across the columns , one result per row

```python
average_per_category = daily_sales.mean(axis=0)
print("Average per category:",average_per_category)

total_per_day= daily_sales.sum(axis=1)
print("Total per day:",total_per_day)

#min per column(worst day per category)
worst_per_category = daily_sales.min(axis = 0)
print("Worst per category:",worst_per_category)

#max per column(best day per category)
best_cat_per_day = daily_sales.max(axis = 1)
print("Best category per day:",best_cat_per_day)
```

### Boolean indexing(Filtering)
Lets you select elements based on a condition . Numpy evaluates the conditiom , createsa True/False array (a mask) and returns only the elements where the condition is true

```python
food_sales = daily_sales[:,0]

#which day had food sales above 10000
mask = food_sales > 13000
print(mask)

# use the mask to get just those values
high_food_days = food_sales[mask]
print(high_food_days)

#shortcut
print(food_sales[food_sales > 13000])

#how many days exceeded the target
print((food_sales > 13000).sum)

#what percentage of days
print((food_sales > 13000).mean() * 100)
```

### Multiple Conditions:(AND) (OR) (NOT)

```python
#AND - both conditions must be true
good_days = daily_sales[(daily_sales[:,0] > 13000) & (daily_sales[:, 1] > 35000)]
print(good_days)

normal_food = food_sales[~(food_sales > 14000)]
print(normal_food)
```

### NP.WHERE : Conditional logic on arrays

```python
#np.where(condion , value if true , value if false)
# Was each day's food sales above 13000? Mark it 'high' or 'low'
food_sales = daily_sales[:,0]
food_label = np.where(food_sales > 13000 ,'high','low')
print(food_label)

#return numeric values
food_flag = np.where(food_sales > 13000 , 1, 0)
print(food_flag)

#keep values above threshold , replace others with the threshold
floor_13k = np.where(food_sales < 13000 , 13000 , food_sales)
print(floor_13k)
```

### Handling Missing Data(NaN)
It is a special float value that means "there is no valid number here" - it is the absence of a value
Any arithmetic NaN return nAn which i why you have to handle it explicitly


```python
#NaN is a float
print(type(np.nan))

#Arithmetic with NaN always returns NaN
print(5 + np.nan)

#NaN is never equal to itself
print(np.nan == np.nan)

#the right way to check for NaN 
print(np.isnan(np.nan))
```

```python
# Inserting and detecting NaN
#Simulate a sales dataset with some missing values 
sales = np.array ([
    [12000,4500,np.nan,22000],
    [np.nan,38000,9500,22000],
    [11000,np.nan,7500,25000],
    [15000,41000,11000,np.nan],
])

#Detecting NaN - RETURNS A TRUE OR A FALSE ARRAY
nan_mask = np.isnan(sales)
print(nan_mask)

#Counting missing values in total
print(f"total missing:{nan_mask.sum()}")

#Counting missing per column(axis = 0)
print("missing per category:",nan_mask.sum(axis = 0))

#Counting missing per row(axis = 1)
print("missing per day:",nan_mask.sum(axis = 1))
```
