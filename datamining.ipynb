### Data Mining Process
**1.Business Understanding**
Business understanding is the process of identifying the business problem and translating it into a data mining problem
What problem are we solving?
Define objectives, goals, success criteria

**2.Data Understanding**
Data understanding involves collecting data, exploring it, and determining whether it is suitable for solving the problem
What data do we have?
Collect, explore, assess data quality

**3.Data preparation**
Data preparation transforms raw data into a clean, organized format suitable for modeling.
Clean, integrate, transform, select features
Is the data ready?

**4.Modeling**
Modeling is the process of selecting, training, and tuning algorithms that learn patterns from the data.
Which algorithm should we use?
Train and tune models

**5.Evaluation**
Evaluation assesses the model's performance using suitable metrics and checks whether it meets the original business objectives.
Does the model meet the objectives?
Test performance, validate results

**6.Deployment**
This is where the model is put into practical use.
How do we use the model?
Integrate, monitor, maintain

### DATA TYPES AND SCALES OF MEASUREMENT
**Nominal**
Categories with no natural order

**Ordinal**
Categories with a natural order but the distances between them are not meaningful or measurable

**Intervals**
Numeric values where the gaps are meaningful but there is no true zero

**Ratio**
Numeric values with meaningful gaps and a true zero meaning zero really means none

### TYPES OF DATA MINING TASKS

#### Descriptive data mining
Descriptive data mining is the process of analyzing data to discover meaningful patterns, relationships, similarities, and summaries without predicting a target variable.

**Goal**
Answer questions such as:
What patterns exist in the data?
Which customers behave similarly?
Which products are frequently bought together?
How can the data be summarized?

**Characteristics**
No target variable (unsupervised learning).
Explores existing data.
Finds hidden structures and relationships.
Helps in decision-making and understanding data.

##### Clustering
Clustering is the process of grouping similar data objects into clusters.
Objects within the same cluster are more similar to each other than to objects in other clusters.

**Applications**
Customer segmentation
Image segmentation
Disease classification
Market research
Social network analysis

**Popular Algorithms**
K-Means
Hierarchical Clustering
DBSCAN

##### Association rule mining
Association rule mining discovers items or events that frequently occur together.

It answers:
"If one item is purchased, what else is likely to be purchased?"

**Applications**
Product recommendations
Store layout planning
Cross-selling
Website recommendations

**Popular Algorithms**
Apriori
FP-Growth

##### Summarization
Summarization creates a compact and meaningful description of a dataset.
Instead of examining every record, it produces an overview

**Applications**
Business reports
Dashboards
Data visualization
Executive summaries

#### Predictive data mining
Predictive data mining uses existing (historical) data to make predictions about future or unknown values.

The dataset contains:
Input variables (features) → information used for prediction
Output variable (target/label) → what we want to predict

Classification predicts a category or class.
Assign an object into one predefined category

Types of Classification
**Binary Classification**
Only two classes.

Alive/Dead
Yes/No

**Multi-class Classification**
More than two classes.

Example
Classifying flowers:

**Multi-label Classification**

**Evaluation Metrics**

##### Regression
Regression predicts a continuous numerical value.
Estimate a quantity.

**Popular Regression Algorithms**
Decision Tree Regression
Random Forest Regression
Support Vector Regression
Linear Regression

**Evaluation Metrics**
Since outputs are numbers, we use:

##### Time series forecasting
Time series forecasting predicts future values using historical observations collected over time.

**What is a Time Series?**
A sequence of observations recorded over time.

Components of a Time Series
*Trend*
Long-term movement.

*Seasonality*
Patterns that repeat regularly.

*Cyclical Patterns*
Long-term cycles not tied to a fixed calendar.

*Random Noise*
Unexpected variation.

Example
Natural disasters.

**Popular Algorithms**
*Classical methods*
ARMA
ARIMA
Moving Average
Exponential Smoothing

*Machine Learning*
Random Forest
Support Vector Regression

*Deep Learning*
Recurrent Neural Networks (RNNs)
Long Short-Term Memory (LSTM)
Gated Recurrent Unit (GRU)
Transformer models

**Evaluation Metrics**
Forecasts are typically evaluated using:
Mean Absolute Error (MAE)
Mean Squared Error (MSE)
Root Mean Squared Error (RMSE)


**Key Difference from Regression**
Unlike ordinary regression, time itself is an important feature because observations are ordered chronologically
Regression often assumes observations are independent.
Time series assumes today's value depends on yesterday's value.

### Where data comes from - ETL vs ELT
# ETL (Extract → Transform → Load) The traditional way
Data is cleaned and structured on a separate, temporary processing server before it is allowed into the target data warehouse

**Process:**
Extract – Collect data from different sources.
Transform – Clean, format, remove duplicates, and prepare the data.Happens at the staging server
Load – Store the processed data in the data warehouse.

**Advantages:**
High-quality data before storage.
Faster reporting and analysis.
Suitable for traditional data warehouses.

**Disadvantages:**
Slower for large datasets.
Original raw data may not be preserved.
Less flexible if new transformations are needed later.

# ELT (Extract → Load → Transform)
Definition:
Data is extracted, loaded as raw data, then transformed inside the data warehouse or data lake.

**Process:**
Extract – Collect raw data.
Load – Store the raw data immediately.
Transform – Clean and process the data when needed.

**Advantages:**
Faster data loading.
Keeps the original raw data.
More flexible and scalable for big data and cloud systems.

**Disadvantages:**
Raw data may contain errors until transformed.
Requires powerful storage and processing systems.


### Databases 
A database is an organized collection of data that is stored electronically and can be easily accessed, managed, and retrieved.

**Why are databases important in data mining?**
Data mining requires large amounts of data to discover patterns and make predictions. Databases:

Store large volumes of data efficiently.
Allow fast searching and retrieval of information.
Keep data organized and secure.
Serve as the main source of data for data mining and machine learning.

**Components of a Database**
Tables – Store data.
Rows (Records) – Represent individual entries.
Columns (Fields/Attributes) – Describe the properties of each record.
Primary Key – A unique identifier for each record.

**Types of Databases**
1. Relational Database (RDBMS)
Stores data in tables (rows and columns).
Tables are related using keys.
Uses SQL (Structured Query Language).

2. NoSQL Database
Stores unstructured or semi-structured data.
Suitable for big data and web applications.

3. Data Warehouse
A central repository that stores historical data from multiple sources.
Optimized for reporting, analytics, and data mining rather than day-to-day transactions.


### DATA PREPROCESSING
Data preprocessing is the process of preparing raw data so that it is clean, consistent, and suitable for data mining or machine learning.

**Steps in Data Preprocessing**
1. Data Cleaning
Remove errors and missing values

2. Data Integration
Combine data from multiple sources

3. Data Transformation
Convert data into a suitable format

4. Data Reduction
Reduce data size while keeping useful information

5. Data Discretization
Convert continuous values into categories

**Benefits of Data Preprocessing**
Improves data quality.
Increases the accuracy of data mining models.
Reduces errors and inconsistencies.
Speeds up data analysis.
Produces more reliable and meaningful results.

**Why is Data Preprocessing Important?**
Real-world data often contains:
Missing values
Duplicate records
Incorrect data
Inconsistent formats
Outliers (extreme values)
Irrelevant information

If these problems are not corrected, the results of data mining or machine learning may be inaccurate.

# Missing Values

Missing values are data fields where no value has been recorded or is unavailable.

They are usually represented by:
Blank cells
NULL
NaN (Not a Number)

N/A

**Types of missing data**
1) MCAR - Missing completely at random
2) MAR - Missing at random
3) MNAR- Missing not at random

**Handling Missing Data**
80% - Data Dropping, drop the whole dataset
30-50% - Mean/Median Imputation ; measures of central tendencies
below 20% - delete rows or columns with missing values

**Causes of Missing Values**
Human error (forgot to enter data)
Equipment or sensor failure
Data transmission errors
Survey respondents skip questions
Information is not applicable
Database integration problems

**Effects of Missing Values**
Reduces data quality.
Produces inaccurate predictions.
Some machine learning algorithms cannot handle missing values.
Can introduce bias into the analysis.

**How to Handle Missing Values**
Delete records (rows)
Delete attributes (columns)
Replace with a constant (e.g., "Unknown")
Replace with the mean (numerical data)
Replace with the median (numerical data with outliers)
Replace with the mode (categorical data)
Predict (impute) the missing values using statistical or machine learning methods

# Duplicate Records
Duplicates are records that appear more than once in a dataset.
They represent the same entity or event multiple times.

**Causes of Duplicates**
Data entered multiple times.
Data imported repeatedly.
Merging databases without checking existing records.
System or software errors.

**Effects of Duplicates**
Increase the size of the dataset unnecessarily.
Produce incorrect statistics.
Bias machine learning models.
Lead to incorrect reports.

**How to Handle Duplicates**
Identify duplicate records using unique identifiers (e.g., Student ID, Customer ID).
Remove exact duplicates.
Merge duplicate records if they contain complementary information.
Validate data before importing it.

# Outliers
An outlier is a value that is significantly different from the rest of the data.
It lies unusually far from the majority of observations

**Causes of Outliers**
Data entry mistakes.
Measurement or sensor errors.
Fraudulent activities.
Genuine but rare events.
Natural variation.

**Effects of Outliers**
Distort averages (means).
Affect regression models.
Reduce prediction accuracy.
Mislead data analysis.

**How to Handle Outliers**
Verify whether the value is an error.
Correct obvious mistakes.
Remove outliers if they are incorrect or irrelevant.
Transform the data (e.g., using logarithms).
Use robust statistics such as the median instead of the mean.
Keep the outlier if it represents a valid and important observation.

**Detecting Outliers**
1. Box Plot
Values outside the whiskers are potential outliers.

2. Z-Score
A value with a large absolute Z-score (often greater than 3 or less than -3) may be considered an outlier.

3. Interquartile Range (IQR)
A value is considered an outlier if it is:
Less than Q1 − 1.5 × IQR, or
Greater than Q3 + 1.5 × IQR

### Anonymity and Pseudonymity

# Anonymity
Anonymity is the state in which a person's identity is completely unknown and cannot be identified from the available data.

Anonymity is the process of removing or hiding all personally identifiable information (PII) so that an individual cannot be identified.

**Characteristics**
Personal identity is removed.
The original person cannot reasonably be identified.
The process is generally irreversible.
Used when organizations want to analyze data without revealing who the individuals are.

**Applications**
Medical research
Government statistics
Census data
Academic research
Public datasets

**Advantages**
Strong protection of personal privacy.
Lowers the risk of identity theft.
Helps organizations share data more safely.

**Disadvantages**
It may be impossible to reconnect the data to the original individual if needed.
Removing too much information can reduce the usefulness of the data for analysis

# Pseudonymity (Pseudonymization)
Pseudonymity is the process of replacing a person's identity with a fake identifier (pseudonym) instead of removing it completely.

Pseudonymity is a technique where identifiable information is replaced with artificial identifiers (such as codes or IDs), while the original identity is stored separately and can be recovered if authorized.

**Characteristics**
Identity is hidden but not permanently removed.
The original identity can be recovered using the separate mapping table.
Often reversible under strict security controls.

**Applications**
Hospitals
Banks
Insurance companies
Clinical trials

**Advantages**
Protects privacy while allowing records to be linked over time.
Makes it possible to update or correct records.
Useful when researchers need to revisit the same individual's data without exposing their identity.

**Disadvantages**
If the mapping table is compromised, identities can be revealed.
Requires strong security measures to protect the key linking pseudonyms to real identities.

### BIAS AND FAIRNESS
# Bias
Bias is a systematic error or unfair preference in data, algorithms, or decision-making that leads to inaccurate or discriminatory outcomes.

Definition: Bias is the tendency of a dataset, model, or process to favor or disadvantage certain individuals or groups unfairly.

Causes of Bias
1. Data Collection Bias
The collected data does not represent the whole population

a. Sampling Bias
The sample chosen for analysis is not representative.

b. Historical Bias
Historical inequalities are reflected in the data.

2. Measurement Bias
The data is measured inaccurately or inconsistently.

Example:
A faulty temperature sensor records incorrect values.

3. Algorithmic Bias
The algorithm itself learns unfair patterns or gives different outcomes to different group

**Effects of Bias**
Unfair hiring decisions.
Discrimination in loan approvals.
Incorrect medical diagnoses.
Biased facial recognition.
Reduced trust in AI systems.

**How to Reduce Bias**
Collect diverse and representative data.
Remove unnecessary sensitive attributes where appropriate.
Balance the dataset.
Test models across different groups.
Regularly monitor and retrain models.

# Fairness
Fairness means ensuring that data mining and machine learning systems treat individuals and groups equitably, without unjust discrimination.

Definition: Fairness is the principle that decisions made by data-driven systems should be impartial, equitable, and free from unjust bias.

**Characteristics of Fair Systems**
Treat similar cases similarly.
Avoid unjust discrimination.
Be transparent where possible.
Produce consistent decisions.
Be regularly monitored for unfair outcomes.

**Importance of Fairness**
Build public trust.
Promote equal opportunities.
Reduce discrimination.
Improve ethical decision-making.
Support compliance with laws and regulations.

**The main subcategories of fairness**
1. Individual Fairness – Similar individuals should be treated similarly.
2. Group Fairness – Different groups should be treated equitably.
3. Demographic Parity – Equal positive outcome rates across groups.
4. Equal Opportunity – Qualified individuals have equal chances of success.
5. Equalized Odds – Similar prediction error rates across groups.
6. Procedural Fairness – The decision-making process is fair and consistent.
7. Outcome (Distributive) Fairness – Outcomes are distributed fairly.
